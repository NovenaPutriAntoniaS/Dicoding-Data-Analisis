### Pertanyaan 1:

````python
#create a linear regression model to predict hourly bike rentals based on temperature
x = hour_data[['temp']]
y = hour_data['cnt']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse:.2f}')
````

### Pertanyaan 2:

````python
#create linear regression model to predict daily bike rentals based on season
x = day_data[['season', 'temp']]
y = day_data['cnt']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse:.2f}')
````

````python
from sklearn.cluster import KMeans

rfm_data = day_data[['temp', 'cnt']]
kmeans = KMeans(n_clusters=3, random_state=42)
rfm_data['cluster'] = kmeans.fit_predict(rfm_data[['temp', 'cnt']])

plt.scatter(rfm_data['temp'], rfm_data['cnt'], c=rfm_data['cluster'], cmap='rainbow')
plt.xlabel('temp')
plt.ylabel('cnt')
plt.title('Customer Segmentation')
plt.show()
````

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your data into the hour_data variable.
# Replace 'your_data.csv' with the actual file path.
hour_data = pd.read_csv('hour.csv')

# Assuming hour_data is your original dataframe
rfm_data = hour_data [['cnt', 'temp', 'season']].copy()

# Define thresholds for clustering
cnt_thresholds = [0, 100, 200, 500]  # adjust these values based on your data
temp_thresholds = [-10, 0, 10, 20]  # adjust these values based on your data

# Create clusters based on thresholds
rfm_data['cluster'] = 0
for i in range(1, len(cnt_thresholds)):
    rfm_data.loc[(rfm_data['cnt'] > cnt_thresholds[i-1]) & (rfm_data['cnt'] <= cnt_thresholds[i]), 'cluster'] = i

for i in range(1, len(temp_thresholds)):
    rfm_data.loc[(rfm_data['temp'] > temp_thresholds[i-1]) & (rfm_data['temp'] <= temp_thresholds[i]), 'cluster'] += 3

# Plot the clusters
sns.scatterplot(x='cnt', y='temp', hue='cluster', data=rfm_data)
plt.title('RFM Analysis')
plt.xlabel('bike rentals')
plt.ylabel('temperature (c)')
plt.show()
````
